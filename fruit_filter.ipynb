```python
# importing libraries 
import cv2 
import numpy as np 
  
image = cv2.imread('/home/kmit/Downloads/images/fruits.jpeg') 
  
cv2.imshow('Original Image', image) 
cv2.waitKey(0) 
  
# Gaussian Blur 
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 
cv2.imshow('Gaussian Blurring', Gaussian) 
cv2.waitKey(0) 
  
# Median Blur 
median = cv2.medianBlur(image, 5) 
cv2.imshow('Median Blurring', median) 
cv2.waitKey(0) 
  
  
# Bilateral Blur 
bilateral = cv2.bilateralFilter(image, 9, 75, 75) 
cv2.imshow('Bilateral Blurring', bilateral) 
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
def gaussian_transform(arr):

    u=arr.mean()

    sig=arr.std()

    fd=[]

    a=0
    pi=math.pi
    for i in arr:

        a=((i-u)**2)/(2*(sig**2))
        e=math.exp(-a)
        fd.append((1/(sig*math.sqrt(2*pi))) * e)


    return fd
```

```python
# import the required PIL Modules

from PIL import Image

from PIL import ImageFilter

 

# Load an image

imageObject = Image.open("/home/kmit/Downloads/images/fruits.jpeg");

plt.subplot(121),plt.imshow(imageObject),plt.title('Original')
plt.xticks([]), plt.yticks([])

# Show the original image

#imageObject.show();

 

# Apply mode filter multiple times

modeFilterApplied = imageObject.filter(ImageFilter.ModeFilter);

for i in range (0, 6):

    modeFilterApplied = modeFilterApplied.filter(ImageFilter.ModeFilter);

 

# Show the image with mode filter applied

#modeFilterApplied.show();

plt.subplot(122),plt.imshow( modeFilterApplied ),plt.title('Mode filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

# Importing Image and ImageFilter module from PIL package   
from PIL import Image, ImageFilter  
     
# creating a image object  
im1 = Image.open("/home/kmit/Downloads/images/rank_image.jpeg");  
     
# applying the rank filter  
im2 = im1.filter(ImageFilter.RankFilter(size = 3, rank = 0))  
im3 = im1.filter(ImageFilter.RankFilter(size = 3, rank = 3 * 3-1)) 
im4 = im1.filter(ImageFilter.RankFilter(size = 3, rank = (3 * 3)//2))
#im2.show() 

plt.subplot(121),plt.imshow(im1),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(im2),plt.title('mini-Mode filter')
plt.xticks([]), plt.yticks([])

```

```python
sub1 = plt.subplot(2, 2, 1)
sub1.imshow(im1),plt.title('Original')
sub1.set_xticks(())
sub1.set_yticks(())
sub2 = plt.subplot(2, 2, 2)
sub2.imshow(im2),plt.title('mini-Mode filter')
sub2.set_xticks(())
sub2.set_yticks(())
sub3 = plt.subplot(2, 2, 3)
sub3.imshow(im3),plt.title('Median-Mode filter')
sub3.set_xticks(())
sub3.set_yticks(())
sub4 = plt.subplot(2, 2, 4)
sub4.imshow(im4),plt.title('maximum-Mode filter')
sub4.set_xticks(())
sub4.set_yticks(())
```

```python
mode_filter=im1.filter(ImageFilter.ModeFilter(size=3))
#mode_filter.show()
plt.subplot(121),plt.imshow(im1),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(mode_filter),plt.title('mini-Mode filter')
plt.xticks([]), plt.yticks([])
```

```python
from scipy import ndimage

varianceMatrix = ndimage.generic_filter(img1, np.var, size = 3)
plt.subplot(121),plt.imshow(img1),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(varianceMatrix),plt.title('variance filter')
plt.xticks([]), plt.yticks([])
```
