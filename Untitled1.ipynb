```python
import numpy
import matplotlib.pyplot as plt
from copy import deepcopy
from PIL import Image
from math import cos, sin

img1 = Image.open('/home/raja/kmit/classes/brain.png')
gray = img1.convert('L')
img = numpy.asarray(gray)
gray1= numpy.asarray(gray)

c = 1
y = float(input("input y :"))

# copy list not reference
pwl = deepcopy(gray1)

min = 99999999999999
max = 0

for i in range(len(img)):
    for j in range(len(img)):
        s = (c*img[i][j])**y
        #s = img[i][j]
        if(s > max):
            max = s
        if(s < min):
            min = s

for i in range(len(img)):
    for j in range(len(img)):
        s = (c*img[i][j])**y
        s = (s-min)/(max-min)
        s = (s*254)+.15
        pwl[i][j] = s 

print('min:',min,'max:',max)

plt.subplot(2, 2, 1)
plt.imshow(gray1)
plt.subplot(2, 2, 2)
plt.imshow(pwl)


plt.show()
```
