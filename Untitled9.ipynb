```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('/home/raja/kmit/classes/face_filter.jpeg')
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# edge_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpen_kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
img = cv2.filter2D(grayscale, -1, sharpen_kernel)

# Smooth out image
# blur = cv2.medianBlur(img, 3)
blur = cv2.GaussianBlur(img, (3,3), 0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
im_ravel=img.ravel()
```

```python

plt.hist(im_ravel,256,[0,256])
```

```python
equalize = cv2.equalizeHist(img)
plt.hist(equalize.ravel(), 256, [0,256])
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
clahe = cv2.createCLAHE().apply(img)
plt.hist(clahe.ravel(), 256, [0,256])
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(equalize),plt.title('equalize')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(clahe),plt.title('clahe')
plt.xticks([]), plt.yticks([])

plt.show()
```
