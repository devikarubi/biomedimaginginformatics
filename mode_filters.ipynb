```python
# Importing Image and ImageFilter module from PIL package   
from PIL import Image, ImageFilter  
     
# creating a image object  
im1 = Image.open("/home/raja/kmit/classes/rank_image.jpeg");  
     
# applying the rank filter  
# mini-mode
im2 = im1.filter(ImageFilter.RankFilter(size = 3, rank = 0))
#median-mode
im3 = im1.filter(ImageFilter.RankFilter(size = 3, rank = 3 * 3-1)) 
#max-mode
im4 = im1.filter(ImageFilter.RankFilter(size = 3, rank = (3 * 3)//2))
#im2.show() 
```

```python
sub1 = plt.subplot(2, 2, 1)
sub1.imshow(im1),plt.title('Original')
sub1.set_xticks(())
sub1.set_yticks(())
sub2 = plt.subplot(2, 2, 2)
sub2.imshow(im2),plt.title('mini-Mode filter')
sub2.set_xticks(())
sub2.set_yticks(())
sub3 = plt.subplot(2, 2, 3)
sub3.imshow(im3),plt.title('Median-Mode filter')
sub3.set_xticks(())
sub3.set_yticks(())
sub4 = plt.subplot(2, 2, 4)
sub4.imshow(im4),plt.title('maximum-Mode filter')
sub4.set_xticks(())
sub4.set_yticks(())
```

```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
%matplotlib inline

image = cv2.imread('/home/raja/kmit/classes/face_filter.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(11,6))
plt.imshow(image)
plt.title('Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```python
new_image = cv2.Laplacian(image2,cv2.CV_64F)

plt.figure(figsize=(11,6))
plt.subplot(131), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(new_image, cmap='gray'),plt.title('Laplacian')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(image2 + new_image, cmap='gray'),plt.title('Resulting image')
plt.xticks([]), plt.yticks([])
plt.show()
```
