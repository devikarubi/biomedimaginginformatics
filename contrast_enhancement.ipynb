```python
import numpy
import matplotlib.pyplot as plt
from copy import deepcopy
from PIL import Image
from math import cos, sin

img = Image.open('/home/raja/kmit/classes/TCGA-CS-4938-01Z-00-DX1_20x_17248_8864_256x256.png')
gray = img.convert('L')
img1 = numpy.asarray(gray)

c = 1
y = float(input("input y :"))

# copy array not reference
pwl = deepcopy(img1)

min = 99999999999999
max = 0

for i in range(len(img1)):
    for j in range(len(img1[i])):
        s = (c*img1[i][j])**y
        if(s > max):
            max = s
        if(s < min):
            min = s
            
for i in range(len(img1)):
    for j in range(len(img1[i])):
        s = (c*img1[i][j])**y
        s = (s-min)/(max-min)
        s = s*254
                
        pwl[i][j] =s    

print('min:',min,'max :',max)

plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.subplot(2, 2, 2)
plt.imshow(pwl)


plt.show()
```
