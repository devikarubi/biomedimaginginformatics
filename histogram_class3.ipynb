```python
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
col_image = Image.open('/home/raja/kmit/classes/TCGA-CS-4938-01Z-00-DX1_20x_17248_8864_256x256.png')
```

```python
col_image
```

```python
#an L mode image, that means it is a single channel image - normally interpreted as greyscale. 
#The L means that is just stores the Luminance
gray = col_image.convert('L')
```

```python
gray
```

```python
#change image as np array
np_gray1=np.array(gray)
```

```python
np_gray1.shape
```

```python
np_gray1[0,0] 
```

```python
#change image as np array
np_col_image=np.array(col_image)
```

```python
np_col_image.shape
```

```python
np_col_image[0,0,:]
```

```python
gray_ravel=np_gray1.ravel()
```

```python
gray_ravel.shape
```

```python
gray_ravel[0]
```

```python
color_ravel=np_col_image.ravel()
```

```python
color_ravel.shape
```

```python
color_ravel[0],color_ravel[1],color_ravel[2]
```

```python
#Histogram for gray image
plt.hist(gray_ravel,5,[0,256]); plt.show()
```

```python
#seperate red ,green and blue channels
col_red=np_col_image[:,:,0]
col_green=np_col_image[:,:,1]
col_blue=np_col_image[:,:,2]
```

```python
col_red.shape,col_green.shape,col_blue.shape
```

```python
col_red[0,0],col_green[0,0],col_blue[0,0]
```

```python
#ravel the RGB channel
red_ravel=col_red.ravel()
green_ravel=col_green.ravel()
blue_ravel=col_blue.ravel()
```

```python
red_ravel.shape, green_ravel.shape, blue_ravel.shape
```

```python
#histogram for color image-channelwise
plt.title("Histogram for RED channel")
plt.hist(red_ravel,256,[0,256]); plt.show()
plt.title("Histogram for GREEN channel")
plt.hist(green_ravel,256,[0,256]); plt.show()
plt.title("Histogram for BLUE channel")
plt.hist(blue_ravel,256,[0,256]); plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#img = cv2.imread('home.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([np_col_image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
