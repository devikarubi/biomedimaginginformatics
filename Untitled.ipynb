```python
for i in range(0, im1.size[0]-1):

        for j in range(0, im1.size[1]-1):

            # Get pixel value at (x,y) position of the image

            f = im1.getpixel((i,j));
            gray_invert=gray(i,j)
            

           

            # Do invert of the pixel

            redPixel    = 255-round(f[0]);

            greenPixel  = 255-round(f[1]);

            bluePixel   =255-round( f[2]);
            gray_intensity=255-round(gray_invert)

 

            # Modify the image with the transformed pixel values

            im2.putpixel((i,j),(redPixel, greenPixel, bluePixel));
            gray_invert.putpixel((i,j),gray_intensity)
            
```
